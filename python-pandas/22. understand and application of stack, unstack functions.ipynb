```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'region': ['seoul', 'incheon', 'gyenggi', 'ulsan', 'busan', 'gangwon', 'daejeon'],
    'day': ['monday','tuesday', 'wednesday','thursay', 'friday', 'saturday', 'sunday'],
    'precipitation': [100, 80, 1000, 200, 200, 100, 200],
    'rainfall probability': [80, 70, 90, 10, 20, 30, 20]
})

df
```

#### stack & unstack
- stack: change dataframe from column level to index level
- unstack: change dataframe from index level to column level
    - opposite of stack operation

```python
new_df = df.set_index(['region', 'day'])
new_df
```

```python
new_df.index
```

```python
new_df.unstack(0)
```

```python
new_df.unstack(1)
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```
