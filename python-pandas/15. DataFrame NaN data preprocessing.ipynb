#### how to deal with NaN
- delete from data
    - use of dropna function 
- replace with other value
    - fillna function
 

#### delete NaN data

```python
train_data.dropna() # deletes a row if contains NaN
```

```python
train_data.dropna(subset=['Age', 'Cabin']) #drops if Age or Cabin is na
```

```python
train_data.dropna(axis=1)
```

#### replace NaN value
- replace by avg value
- replace by death/survivor avg value

```python
train_data['Age'].fillna(train_data['Age'].mean())
```

```python
# average of survivor age
mean1 = train_data[train_data['Survived'] == 1]['Age'].mean()

# average of death age
mean0 = train_data[train_data['Survived'] == 0]['Age'].mean()
print (mean1, mean0)
```

```python
train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```

```python
# replace the NaN values from original train_data
train_data.loc[train_data['Survived'] == 1, 'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data.loc[train_data['Survived'] == 0, 'Age'] = train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```
