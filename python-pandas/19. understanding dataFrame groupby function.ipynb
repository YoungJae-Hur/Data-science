```python
import pandas as pd
import numpy as np
```

#### group by
- apply the 3 steps and grouping (similar as sql group by)
    - 1. data partition
    - 2. apply operation
    - 3. merge data

```python
train_data = pd.read_csv('train.csv')
train_data.head()
```

#### GroupBy groups attribute
- express each group's index into dict shape

```python
class_group = train_data.groupby('Pclass')
class_group
```

```python
class_group.groups
```

```python
gender_group = train_data.groupby('Sex')
gender_group
```

```python
gender_group.groups
```

#### grouping function
- appliable statistics function
- count: num of data
- sum: sum of data
- mean, std, var: avg, standard deviation, distribution
- min, max: 

```python
class_group.count()
```

```python
class_group.sum() # no big deal...
```

```python
class_group.mean()['Age']
```

```python
class_group.mean()['Survived']
```

```python
class_group.min()
```

#### calculate survivor rate for each sex

```python
train_data.groupby('Sex').mean()
```

```python
train_data.groupby('Sex').mean()['Survived']
```

#### grouping w/ multiple columns
- feed column list in groupby param
- result after applied statistics is in form of multi-index dataframe


#### calculating survivor rate based on class and sex

```python
train_data.groupby(['Pclass', 'Sex']).mean()
```

```python
train_data.groupby(['Pclass', 'Sex']).mean().loc[(2,'female')]
```

```python
train_data.groupby(['Pclass', 'Sex']).mean()['Survived']
```

```python
train_data.groupby(['Pclass', 'Sex']).mean().index
```

#### group by using index
- if index exists, available to use level in groupby function
    - *level means depth of index

- set_index function
    - change column data to index level
- reset_index function
    - resets index value

```python
train_data.head()
```

```python
train_data.set_index('Pclass')
```

```python
train_data.set_index(['Pclass', 'Sex'])
```

```python
train_data.set_index(['Pclass', 'Sex']).reset_index()
```

```python
train_data.set_index('Age').groupby(level=0).mean()
```

### calculating survivor rate by ages

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

```python
train_data.set_index('Age').groupby(age_categorize).mean()
```

```python
train_data.set_index('Age').groupby(age_categorize).mean()['Survived']
```

#### grouping using multi-index

```python
train_data.set_index(['Pclass', 'Age']).groupby(level=[0,1]).mean()
```

```python
train_data.set_index(['Pclass', 'Sex']).groupby(level=[0,1]).mean()
```

#### using aggregate (집계) function 
- available to check by each groups applying aggregate function for groupby result 

```python
train_data.set_index(['Pclass', 'Sex']).groupby(level=[0,1]).aggregate([np.mean, np.sum, np.max])
```
