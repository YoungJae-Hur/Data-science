```python
import pandas as pd
import numpy as np
```

#### merge dataframe with using concat function
- pandas concat function
- can merge based on axis

```python
df1 = pd.DataFrame({'key1':np.arange(10), 'value1': np.random.randn(10)})
```

```python
df2 = pd.DataFrame({'key1':np.arange(10), 'value1': np.random.randn(10)})
```

```python
df2
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

#### when column name is different

```python
df3= pd.DataFrame({'key2': np.arange(10), 'value2':np.random.randn(10)})
```

```python
pd.concat([df1, df3])
```

```python
pd.concat([df1, df3], axis=1)
```
