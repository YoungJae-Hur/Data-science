```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'region': ['seoul', 'incheon', 'gyenggi', 'ulsan', 'busan', 'gangwon', 'daejeon'],
    'day': ['monday','tuesday', 'wednesday','thursay', 'friday', 'saturday', 'sunday'],
    'precipitation': [100, 80, 1000, 200, 200, 100, 200],
    'rainfall probability': [80, 70, 90, 10, 20, 30, 20]
})

df
```

#### pivot 
- change the shape of dataframe
- list column for using index, column, data

```python
df.pivot('region', 'day')
```

```python
df.pivot('day', 'region', 'precipitation')
```

```python
# allow duplication after calling
df.pivot('day', 'region')
```

#### pivot_table
- technically same as pivot
- diff b/w pivot
    - if duplication exist, fill out with aggregation function

```python
pd.pivot_table(df, index='day', columns='region', aggfunc=np.mean)
```
