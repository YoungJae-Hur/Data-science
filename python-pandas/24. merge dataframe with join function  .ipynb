```python
import numpy as np
import pandas as pd
```

#### dataframe merge
- merge based on particular column like SQL join
    - join: specify how param
        - inner: default,
        - lef: left outer join,
        - right: right outer join,
        - outer: full outer join,
- pandas.merge is being used

```python
customer = pd.DataFrame({'customer_id' : np.arange(6),
                        'name' : ['brian', 'charlie', 'omega', 'natalie', 'linda', 'jenny'],
                        'age': [28, 35, 46, 29, 32, 34]})
customer
```

```python
orders = pd.DataFrame({'customer_id' : [1,1,2,2,2,3,3,1,4,9],
                       'item' : ['toothbrush', 'toothpaste', 'towel', 'earbuds', 'headphones', 'water', 'italytowel', 'bottledwater', 'case', 'shampoo'],
                       'quantity' : [4,5,7,4,4,3,4,5,6,10]
                      })
orders
```

#### be sure to list which column will be the base column
- on param

```python
pd.merge(customer, orders, on='customer_id', how='inner') # how is the param that equals
```

```python
pd.merge(customer, orders, on='customer_id', how='left')
```

```python
pd.merge(customer, orders, on='customer_id', how='right')
```

```python
pd.merge(customer, orders, on='customer_id', how='outer')
```

```python
cust1 = customer.set_index('customer_id')
```

```python
order1 = orders.set_index('customer_id')
```

```python
cust1
```

```python
order1
```

```python
pd.merge(cust1, order1, left_index=True, right_index=True)
```

#### practice,
- 1. the most sold item?
- 2. what's the most bought item from charlie?


```python
pd.merge(customer, orders, on='customer_id').groupby('item').sum()
```

```python
pd.merge(customer, orders, on='customer_id').groupby('item').sum().sort_values(by='quantity', ascending=False)
```

- 2. the most bought item from charlie

```python
pd.merge(customer, orders, on='customer_id').groupby(['name', 'item']).sum()
```

```python
pd.merge(customer, orders, on='customer_id').groupby(['name', 'item']).sum().loc['charlie', 'quantity']
```
