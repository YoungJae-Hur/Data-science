### # manage data, using dataframe numerical calculation per columns
    - num of clicks (클릭수)
    - rate of clicks (클릭률)
    - avg. cost of clicks (평균클릭비용)

```python
df.head()
```

```python
# rounding off the integer in clicks, apply round function
clk=round(df['클릭수'],0)
clk
```

```python
# remove trailing zero = float -> int 
clk.astype(int)
```

```python
# replace w/ orginal column w/ modified values
df['클릭수'] = clk.astype(int)
df.head()
```

### - rate of clicks : CTR = num of clicks / num of exposures * 100

```python
# populate numerical calculation using columns in datagframe 
df['클릭수']/df['노출수']*100
```

```python
df['클릭률(%)'] = df['클릭수']/df['노출수']*100
df.head()
```

```python
# CPC value (평균클릭비용)
cpc = round(df['평균클릭비용(VAT포함,원)'],0)
df['평균클릭비용(VAT포함,원)'] = cpc.astype(int)
df.head()
```
