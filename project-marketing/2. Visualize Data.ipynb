```python
import matplotlib.pyplot as plt
import pandas as pd
# from pandas import DataFrame
```

```python
df = pd.read_excel('Naver_Report.xls', skiprows=[0])
df.head()
```

```python
dict_data = {'apple': [1,2,3,4], 'strawberry': [2,3,4,5], 'pineapple': [3,4,5,6], 'goldkiwi': [4,5,6,7]}
df = pd.DataFrame(dict_data)
df
```

### # draw chart

```python
# line graph
df.plot()
plt.show()
```

```python
# bar graph
df.plot.bar()
plt.show()
```

```python
# horizontal axis bar
df.plot.barh()
plt.show()
```

```python
# histogram
df.plot.hist()
plt.show()
```

```python
# setting sections in histogram 
df.plot.hist(bins=range(1,9,1))
plt.show()
```

### # adding options in chart

```python
# base bar graph
df.plot.bar()
plt.show()
```

```python
# setting graph size
df.plot.bar(figsize=[10,6]) # 10 is width size(x-axis), 6 is height(y-size)
plt.show()
```

```python
# setting title
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits')
plt.show()
```

```python
# setting size of title font
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits', fontsize=18)
plt.show()
```

```python
# setting name of x-axis
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits', fontsize=18)
plt.xlabel('Number of indices', fontsize=16)
plt.show()
```

```python
# setting name of y-axis
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits', fontsize=18)
plt.xlabel('Number of indices', fontsize=16)
plt.ylabel('Number of fruits', fontsize=16)
plt.show()
```

```python
# setting x-axis gradation
# gradation location, name, fontsize, and angle
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits', fontsize=18)
plt.xlabel('Number of indices', fontsize=16)
plt.ylabel('Number of fruits', fontsize=16)
plt.xticks([0,1,2,3,4], ['first','second','third', 'fourth'], fontsize=10, rotation=0)
plt.show()
```

```python
# setting y-axis gradation
# gradation location, name, fontsize, and angle
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits', fontsize=18)
plt.xlabel('Number of indices', fontsize=16)
plt.ylabel('Number of fruits', fontsize=16)
plt.xticks([0,1,2,3,4], ['first','second','third', 'fourth'], fontsize=10, rotation=0)
plt.yticks([1,3,5,7], ['first','third','fifth', 'seventh'], fontsize=10, rotation=0)
plt.show()
```

```python
# setting a-axis range
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits', fontsize=18)
plt.xlabel('Number of indices', fontsize=16)
plt.ylabel('Number of fruits', fontsize=16)
plt.xticks([0,1,2,3,4], ['first','second','third', 'fourth'], fontsize=10, rotation=0)
plt.yticks([1,3,5,7], ['first','third','fifth', 'seventh'])
plt.xlim([-1, 4])
plt.show()
```

```python
# setting y-axis range
df.plot.bar(figsize=[10,6])
plt.title('Number of fruits', fontsize=18)
plt.xlabel('Number of indices', fontsize=16)
plt.ylabel('Number of fruits', fontsize=16)
plt.xticks([0,1,2,3,4], ['first','second','third', 'fourth'], fontsize=10, rotation=0)
plt.yticks([1,3,5,7], ['first','third','fifth', 'seventh'])
plt.xlim([-1, 4])
plt.ylim([-1, 8])
plt.show()
```
