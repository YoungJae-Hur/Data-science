### extract keywords for management
    - top 5% of num of exposures and num of clicks
        1. find 95percentile (quantile function)
        2. extract 95percentile (top 5 percent)
        3. extract keyword that belongs top 5 percent

```python
df = pd.read_excel('Naver_Report.xls', skiprows=[0])
df.head()
```

```python
# num of exposures 
imp = df['노출수']
```

```python
#1. finding 95percentile
# quantile - print percentile (default : 2nd percentile (median))
imp.quantile()
```

```python
#median
imp.median()
```

```python
# quatile(0) - min number
print(imp.quantile(0))
print(imp.min())
```

```python
# quantile(1) - max number
print(imp.quantile(1))
print(imp.max())
```

```python
# 95 percentile = quantile(0.95)
imp.quantile(0.95)
```

```python
# 25 percentile = quantile(0.25)
print(imp.quantile(0.25))
print(imp.quantile(0.5))
print(imp.quantile(0.75))
print(imp.quantile(0.95))
```

```python
#2. extract 95percentile of num of exposures
#series[condition]
imp=imp[imp>=imp.quantile(0.95)]
imp
```

3. extract keyword that belongs top 5 percent
    
    3.1 extract base top num of exposures
        A. as-is: df base number index
        B. to-be: df reset index to keyword

```python
df_index=df.set_index('키워드')
df_index
```

```python
# imp variable creation or quantile(0.95) df_index 
imp=df_index['노출수']
```

```python
# prints imp, extract top 5%
imp = imp[imp>=imp.quantile(0.95)]
imp
```

```python
# extract top 5% of num of clicks from df_index
clk= df_index['클릭수']
clk=clk[clk>=clk.quantile(0.95)]
clk
```

### # selection of keywords
    - num of cliks and exposures are above 95 percentile

### if more than a condition
    - df[(condition)&(condition)]

```python
#creation of imp, clk variables, df_index
imp=df_index['노출수']
clk=df_index['클릭수']
```

```python
#num of exposure and num of clicks 
result = df_index[(imp>=imp.quantile(0.95))&(clk>=clk.quantile(0.95))]
result
```

```python
#index - prints keyword for management
result.index
```

### # finding low efficient keywords

- imp, cliks high cost despite of being spending cost
    1. below 95 percentile of imps,
    2. below 95 percentile of clks,
    3. above total cost of 85 percentile,
    4. below total cost of 95 percentile

```python
# create cost variable w/ df_index
cost = df_index['총비용(VAT포함,원)']
```

```python
# dateframe[(condition)&(condition)]
result = df_index[(imp<imp.quantile(0.95))&(clk<clk.quantile(0.95))&(cost<cost.quantile(0.95))&(cost>=cost.quantile(0.85))]
result
```
