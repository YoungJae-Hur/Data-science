### data preprocessing based on ad group

```python
#sum of group data 
grouped.sum()
```

```python
#df_group variable
df_group=grouped.sum()
df_group
```

```python
# ctr(percentage of clicks) = clicks / exposures
df_group['클릭률(%)']=df_group['클릭수']/df_group['노출수']
```

```python
# cpc (cost prices of clicks) = total cost / clicks
df_group['평균클릭비용(VAT포함,원)']=df_group['총비용(VAT포함,원)']/df_group['클릭수']
```

```python
# clicks columns of 0 => columns in avg cost of clicks exists w/ null 
df_group[df_group['클릭수'] == 0]
```

```python
#fillna - replaces null values with returned values
df_group['평균클릭비용(VAT포함,원)']=df_group['평균클릭비용(VAT포함,원)'].fillna(0)
```

```python
df_group[df_group['평균클릭비용(VAT포함,원)']==0]
```

```python
# average cost of clicks round(round), removes trailing numbs astype(int)
df_group['평균클릭비용(VAT포함,원)']=round(df_group['평균클릭비용(VAT포함,원)'],0)
df_group['평균클릭비용(VAT포함,원)']=df_group['평균클릭비용(VAT포함,원)'].astype(int)
```

### priority control ad group/low efficient ad group

- priority control ad group
    1. top 20% (above 80percentile of exposures)
    2. top 10% (above 90percentile of clicks)

```python
#create imp, and clicks in df_group 
imp=df_group['노출수']
clk=df_group['클릭수']
```

```python
# if more than one condition
result = df_group[(imp>=imp.quantile(0.8))&(clk>=clk.quantile(0.9))]
```

```python
#index - prints out priority control ad group
result.index
```

- low efficient ad group: imp, clicks that are not top keywords but high cost 
    1. below 80perc of imp
    2. below 90perc of clk
    3. above 60perc of totalcost
    4. below 90perc of totalcost

```python
#create cost variable in df_group
cost=df_group['총비용(VAT포함,원)']
```

```python
#if several conditions
result=df_group[(imp<imp.quantile(0.8))&(clk<clk.quantile(0.9))&(cost>=cost.quantile(0.6))&(cost<cost.quantile(0.9))]
```
