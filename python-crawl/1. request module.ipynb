```python
import requests
```

get 요청하기
    1. http get 요청
    2. query parameter 이용해서 데이터 전달

```python
url = 'https://news.v.daum.net/v/20200414235051213'
resp = requests.get(url)

resp.text
```

post 요청하기
    1.http post 요청하기
    2. post data 이용하여 데이터 전달하기

```python
url = 'https://www.kangcom.com/member/member_check.asp'
data = {
    'id': 'macmath22',
    'pwd': 'Test1357!'
}

resp = requests.post(url, data=data)
resp.text
```

http header 데이터 이용하기 
1. header 데이터 구성하기
2. header 데이터 전달하기 

```python
url = 'https://news.v.daum.net/v/20200414235051213'
headers = {
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.92 Safari/537.36'
}
resp = requests.get(url, headers=headers)
resp.text
```

HTTP response 처리하기
    1. response 객체의 이해
    2. status_code 확인하기
    3. text 속성 확인하기 

```python
url = 'https://news.v.daum.net/v/20200414235051213'
resp = requests.get(url)
if resp.status_code == 200:
    resp.text
else:
    print('error')
```
