1. draw graph using matplotlib module

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

그래프 데이터 생성

```python
x = np.linspace(0, 10, 11)
y = x ** 2 + x + 2 + np.random.randn(11)

print(x)
print(y)
```

#### print a graph
- plot, scatter, hist 
- possible to change using w/ parameter or praph shape or settings w/ plt 

```python
plt.plot(x, y)
```

```python
plt.scatter(x,y)
```

#### adding comments from the graph
- x, y axis or title

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.plot(x, y)
```

### add grid

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)
plt.plot(x, y)
```

### Fix x,y axis range 

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)

plt.xlim(0, 20)
plt.ylim(0, 200)

plt.plot(x, y)
```

### plot param

- change color


```python
plt.plot(x, y, 'k')
```

```python
plt.plot(x, y, '#ff00ff')
```

- change the style of graph

```python
plt.plot(x,y,'-.')
```

```python
plt.plot(x,y,'g^')
```

```python
plt.plot(x,y,'m:')
```

- change the thickness of graph

```python
plt.plot(x,y,'m:', linewidth=9)
```

```python
plt.plot(x,y,color='black', 
         linestyle='--', marker='^', markerfacecolor='blue', markersize=10)
```

#### using subplot, print more than a graph

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'g')

plt.subplot(2,2,3)
plt.plot(x,y,'k')

plt.subplot(2,2,4)
plt.plot(x,np.exp(x),'b')

```

### hist 
- create histogram
- sets the histogram bar using bins

```python
data = np.random.randint(1,100, size=200)
print(data)
```

```python
plt.hist(data, bins=20, alpha = 0.5)
plt.xlabel('data')
plt.ylabel('number')
plt.grid(True)
```
